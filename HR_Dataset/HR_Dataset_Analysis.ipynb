Gender Distribution: Analyze the proportion of male vs. female employees.

```python
df.columns
```

```python
df.Sex.value_counts().plot(kind='bar',color=['green','purple'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("proportion of male vs. female employees")
```

Marital Status Distribution: Examine the marital status of employees to understand the diversity of marital backgrounds.

```python
df.columns
```

```python
df[['MarriedID', 'MaritalStatusID','MaritalDesc']]
```

```python
df.MarriedID.unique(),df.MaritalStatusID.unique(),df.MaritalDesc.unique()
```

```python
sns.countplot(x='MaritalDesc',data = df,color='blue')
plt.xlabel('Marital Status')
plt.ylabel("Count")
plt.title("Marital status of employees")
plt.show()
```

Race and Ethnicity Distribution: Assess the racial and ethnic composition of the workforce to gauge diversity and inclusion efforts.

```python
df.columns
```

```python
df['HispanicLatino']=df['HispanicLatino'].str.replace('No','no').str.replace('Yes','yes')
```

```python
df['HispanicLatino'].value_counts()
```

```python
sns.countplot(x='RaceDesc',hue='HispanicLatino',data =df,palette='rocket')
plt.xticks(rotation=90)
plt.xlabel("Race Description")
plt.ylabel("Count")
plt.title("Race and Ethnicity Distribution")
plt.show()
```

```python
df['RaceDesc'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Race and Ethnicity Distribution')
plt.ylabel('')
plt.show()
```

### 2. Employee Satisfaction and Engagement

Employee Satisfaction

```python
df.columns
```

```python
df.EmpSatisfaction.value_counts()
```

```python
sns.countplot(x='EmpSatisfaction',hue='Department',data = df)
plt.xlabel("Satisfaction Rating")
plt.ylabel('Count')
plt.title("Department wise Satisfaction of Employees")
plt.show()
```

```python
sns.countplot(x='EmpSatisfaction',hue='Sex',data = df)
plt.xlabel("Satisfaction Rating")
plt.ylabel('Count')
plt.title("Gender wise Satisfaction of Employees")
plt.show()
```

```python
sns.countplot(x='EmpSatisfaction',hue='CitizenDesc',data = df)
plt.xlabel("Satisfaction Rating")
plt.ylabel('Count')
plt.title("Citizen wise Satisfaction of Employees")
plt.show()
```

```python
sns.countplot(x='EmpSatisfaction',hue='RaceDesc',data = df)
plt.xlabel("Satisfaction Rating")
plt.ylabel('Count')
plt.title("Citizen wise Satisfaction of Employees")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['EmpSatisfaction'], kde=True)
plt.title('Employee Satisfaction Levels')
plt.xlabel('Satisfaction Score')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='PerfScoreID', y='PerformanceScore', data=df)
plt.title('Performance Scores by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Performance Score')
plt.show()
```
